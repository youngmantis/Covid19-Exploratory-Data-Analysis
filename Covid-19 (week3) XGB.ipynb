```python
all_data = pd.concat([train,test])
all_data
```

```python
ntrain = train.shape[0]
ntest = test.shape[0]
```

```python
y_train_c = train.ConfirmedCases.values.astype(int)
```

```python
y_train_ft = train.Fatalities.values.astype(int)
```

```python
cols = ['ConfirmedCases'],['Fatalities']
```

```python
all_data.drop(['ConfirmedCases', 'Fatalities'], axis = 1, inplace = True)
```

```python
all_data
```

```python
all_data = pd.get_dummies(all_data)
print(all_data.shape)
```

```python
x_train = all_data[:ntrain]
x_test = all_data[:ntrain]
```

```python
x_train.shape, x_test.shape, y_train_c.shape, y_train_ft.shape
```

```python
cXGB = XGBRegressor()
cXGB.fit(x_train, y_train_c)
```

```python
y_pred_c = cXGB.predict(x_test)
y_pred_c = y_pred_c.astype(int)
y_pred_c[y_pred_c <0]=0
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
predictions = cXGB.predict(x_test)
mae = mean_absolute_error(predictions, y_pred_c)
print("Mean Absolute Error: {:,.0f}".format(mae))
```

```python
fXGB = XGBRegressor()
fXGB.fit(x_train, y_train_ft)
```

```python
y_pred_ft = fXGB.predict(x_test)
y_pred_ft = y_pred_ft.astype(int)
y_pred_ft[y_pred_ft <0]=0
```

```python
predictions1 = fXGB.predict(x_test)
mae = mean_absolute_error(predictions1, y_pred_ft)
print("Mean Absolute Error: {:,.0f}".format(mae))
```

```python
submission = pd.DataFrame({'ForecastId': test.index,'ConfirmedCases':predictions,'Fatalities':predictions1})
filename = 'submission.csv'

submission.to_csv(filename,index=False)
```
